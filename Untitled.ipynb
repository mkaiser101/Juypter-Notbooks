```python
# use a get requests
html = requests.get("https://www.yellowpages.com/search?search_terms=coffee&geo_location_terms=Los+Angeles%2C+CA")
soup = BeautifulSoup(html.content, "html.parser")

# parse through markup content
all_links = soup.find_all("a")

# loop through each link
# for link in soup.find_all("a"):
#     print "<a href='%s'>%s</a>" %(link.get("href"), link.text)

# parse through all classes defined as "info"
g_data = soup.find_all("div", {"class":"info"})


counter = 1
for item in g_data:
    print  ("%d." %counter)
    print (item.contents[0].find_all("a", {"class":"business-name"})[0].text)
    # use try and except because the list index is out of range
    try:
        print (item.contents[1].find_all("span", {"itemprop":"streetAddress"})[0].text)
    except:
        pass
    try:
        print (item.contents[1].find_all("span", {"itemprop":"addressLocality"})[0].text)
    except:
        pass
    try:
        print (item.contents[1].find_all("span", {"itemprop":"addressRegion"})[0].text)
    except:
        pass
    try:
        print (item.contents[1].find_all("span", {"itemprop":"postalCode"})[0].text)
    except:
        pass
    try:
        print (item.contents[1].find_all("li", {"class":"primary"})[0].text)
    except:
        pass
    try:
        print (item.contents[1].find_all("div", {"itemprop":"telephone"})[0].text)
    except:
        pass
    counter = counter + 1


# below will print the first element
# print item.contents[0].text
```
